# **1. Read and display**
* Read image from file
* Display to screen

# **2. Grayscale**
* Read Image
* Apply grayscale
* Display

```python
# Reading the colored image
image_path = '/kaggle/input/images/2x2_jaafar_omar.jpg'
colored_image = cv2.imread(image_path)

# Convert the colored image to grayscale
grayscale_image = cv2.cvtColor(colored_image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image without axis
plt.axis('off')
plt.imshow(grayscale_image, cmap='gray')
plt.show()
```

# **3. Save to file**
* Read Image
* Apply grayscale
* Save file

```python
# Reading the colored image
image_path = '/kaggle/input/images/2x2_jaafar_omar.jpg'
colored_image = cv2.imread(image_path)

# Convert the colored image to grayscale
grayscale_image = cv2.cvtColor(colored_image, cv2.COLOR_BGR2GRAY)

# Save file to the /kaggle/working directory
output_name = '/kaggle/working/grayscale_image.jpg'
cv2.imwrite(output_name, grayscale_image)

if os.path.exists(output_name):
    print(f"Grayscale image saved as {output_name}")
else:
    print("Failed to save the grayscale image")
```

# **4. Thresholding**
* Read grayscale image
* Apply simple binary thresholding (threshold = 125)
* Display

```python
# Read grayscale image
grayscale_image = cv2.imread('/kaggle/working/grayscale_image.jpg', cv2.IMREAD_GRAYSCALE)

# Apply thresholding
threshold_value = 125
_, thresholded_image = cv2.threshold(grayscale_image, threshold_value, 255, cv2.THRESH_BINARY)

# Display thresholded image
plt.imshow(thresholded_image, cmap='gray')
plt.axis('off')
plt.show()
```

# **6. Histogram**
* Read the grayscale image
* Count the number of pixels for each value (0 to 255)
* Display result

```python
# Read grayscale image
grayscale_image = cv2.imread('/kaggle/working/grayscale_image.jpg', cv2.IMREAD_GRAYSCALE)

# Get the dimensions of the image
height, width = grayscale_image.shape

# Calculate the total number of pixels
total_pixels = height * width

print(f"Total number of pixels in the grayscale image: {total_pixels}\n")

# Calculate the histogram of pixel values
hist = cv2.calcHist([grayscale_image], [0], None, [256], [0, 256])

# Convert Hist to NumPy
hist = np.squeeze(hist)

# Display the count of pixels for each pixel value
for i, count in enumerate(hist):
    print(f"Pixel Value {i}: {int(count)} pixels")
```

# **7. More Histogram**
* repeat #6 using the colored image, count the number of pixels of R, G, &B

```python
# Read the colored image
colored_image = cv2.imread('/kaggle/input/images/2x2_jaafar_omar.jpg')

# Split the image into its color channels
b, g, r = cv2.split(colored_image)

# Calculate the total number of pixels in each channel
total_pixels_b = (b > 0).sum()
total_pixels_g = (g > 0).sum()
total_pixels_r = (r > 0).sum()

print(f"Total number of pixels in the Blue channel: {total_pixels_b}")
print(f"Total number of pixels in the Green channel: {total_pixels_g}")
print(f"Total number of pixels in the Red channel: {total_pixels_r}")
```
